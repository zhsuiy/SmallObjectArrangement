```python
# read info about object orientation
object_orientation = {}

with open('ParentObjFrontDirection.txt') as fp:
    flag = False   
    mapping={}
    sceneid = ''
    for line in fp:        
        if line.strip().endswith('.txt'): # new scene
            sceneid = line.split('.')[0]
            #print sceneid
            mapping = {}
            flag = True           
        elif flag == True: #read mapping
            if (len(line.strip()) == 0):
                flag = False
                object_orientation[sceneid] = mapping
                #print('test')
            info = line.strip().split('|')
            if len(info) > 1: # skip room
                mapping[info[0].split('\t')[0]] = info[1]        

print len(object_orientation)
        
```

```python
# read scene info
from os import walk
scenepath='./../scenes'

f = []
for (dirpath, dirnames, filenames) in walk(scenepath):      
    f.extend(filenames)
    break

sceneinfo = {}
for f in filenames:    
    if f.endswith('.txt'):            
        # get all model index mapping prent contact position
        # for each scene
        sceneid = f.split('.')[0]
        modelindex_parentcontact_mapping = {}
        modelindex_name_mapping = {}
        modelindex_children_mapping = {}
        with open(scenepath + '/' + f) as fp:
            for line in fp:
                if line.startswith('newModel'):
                    modelid = line.strip().split(' ')[-1]
                    modelIndex = line.strip().split(' ')[1]
                    modelindex_name_mapping[modelIndex] = modelid
                if line.startswith('parentContactPosition'):
                    modelindex_parentcontact_mapping[modelIndex] = line.strip().split(' ')[1:]
                if line.startswith('children'):
                    childrenids = line.strip().split(' ')
                    if len(childrenids) > 1 and modelIndex != '0': # this is a parent, we need to record the order infomation
                        modelindex_children_mapping[modelIndex] = childrenids[1:]
        #print modelindex_parentcontact_mapping
        #print modelindex_name_mapping
        #print (modelindex_children_mapping)
        mapping = {}
        mapping['modelindex_parentcontact_mapping'] = modelindex_parentcontact_mapping
        mapping['modelindex_name_mapping'] = modelindex_name_mapping
        mapping['modelindex_children_mapping'] = modelindex_children_mapping
        sceneinfo[sceneid] = mapping
        
        
           
       
```

```python
# calculate pairwise information
height_higher_preferences = {}
height_equal_preferences = {}
depth_front_preferences = {}
depth_equal_preferences = {}
medium_center_preferences = {}
medium_equal_preferences = {}

HG = 'AHigherB'
HE = 'AHighEqualB'

MG = 'AMoreMediumB'
ME = 'AEqualMediumB'

ZG = 'AFrontB'
ZE = 'ADepthEqualB'


for sceneid in sceneinfo:
    #print sceneid
    modelindex_parentcontact_mapping = sceneinfo[sceneid]['modelindex_parentcontact_mapping']
    modelindex_name_mapping = sceneinfo[sceneid]['modelindex_name_mapping']
    modelindex_children_mapping = sceneinfo[sceneid]['modelindex_children_mapping']
    
    orientations = object_orientation[sceneid]
    for parentindex in modelindex_children_mapping:
        orient = orientations[parentindex]
        p_contact_pos = modelindex_parentcontact_mapping[parentindex]
        p_x = float(p_contact_pos[0])
        p_z = float(p_contact_pos[1])
        allchildren = modelindex_children_mapping[parentindex]
        children_in_height = {}
        for c in allchildren:
            height = float(modelindex_parentcontact_mapping[c][2])
            if height not in children_in_height:
                children_in_height[height] = [c]
            else:
                children_in_height[height].append(c)
        
        # handle height info       
        height_list = sorted(children_in_height.keys())  
        for i in range(len(height_list)):
            for j in range(i+1, len(height_list)):
                lower_children = children_in_height[height_list[i]]
                higher_children = children_in_height[height_list[j]]
                for lc in lower_children:
                    for hc in higher_children:
                        lcid = modelindex_name_mapping[lc]
                        hcid = modelindex_name_mapping[hc]
                        if lcid in model_id_cat_map and hcid in model_id_cat_map:
                            lccat = model_id_cat_map[lcid]
                            hccat = model_id_cat_map[hcid]
                            if lccat == hccat:                        
                                continue
                            # order matters, the first one is higher than the second one
                            pair = (hccat,lccat)
                            if pair not in height_higher_preferences:
                                height_higher_preferences[pair] = 1
                            else:
                                height_higher_preferences[pair] += 1
        
                
        # analyze the relationship within the same height
        for h in children_in_height:
            children = children_in_height[h]
            for i in range(len(children)):
                for j in range(i+1,len(children)):
                    c1 = children[i]
                    c2 = children[j]
                    c1id = modelindex_name_mapping[c1]
                    c2id = modelindex_name_mapping[c2]

                    if c1id in model_id_cat_map and c2id in model_id_cat_map:                    
                        c1cat = model_id_cat_map[c1id]
                        c2cat = model_id_cat_map[c2id]
                        if c1cat == c2cat:                        
                            continue
                        
                        # save same height information
                        # make sure the pair is of same order of same categories, order doesn't matter
                        pair =(c1cat,c2cat) if c1cat < c2cat else (c2cat,c1cat)                        
                        if pair not in height_equal_preferences:                            
                            height_equal_preferences[pair] = 1
                        else:
                            height_equal_preferences[pair] += 1
                        
                        cat_x_center = ''
                        cat_x_side = ''
                        flag_x_equal = False
                        
                        cat_z_front = ''
                        cat_z_back = ''
                        flag_z_equal = False
                        
                        
                        c1_contact_pos = modelindex_parentcontact_mapping[c1]
                        c1_x = float(c1_contact_pos[0])
                        c1_z = float(c1_contact_pos[1])
                        c2_contact_pos = modelindex_parentcontact_mapping[c2]
                        c2_x = float(c2_contact_pos[0])
                        c2_z = float(c2_contact_pos[1])
                        
                        if orient == 'ToBack':
                            deltax1 = abs(c1_x - p_x)
                            deltax2 = abs(c2_x - p_x)
                            if deltax1 < deltax2:
                                cat_x_center = c1cat
                                cat_x_side = c2cat
                            elif deltax2 < deltax1:
                                cat_x_center = c2cat
                                cat_x_side = c1cat
                            else:
                                flag_x_equal = True
                                
                            if c1_z < c2_z:
                                cat_z_front = c1cat
                                cat_z_back = c2cat
                            elif c2_z < c1_z:
                                cat_z_front = c2cat
                                cat_z_back = c1cat
                            else:
                                flag_z_equal = True
                            
                        elif orient == 'ToFront':
                            deltax1 = abs(c1_x - p_x)
                            deltax2 = abs(c2_x - p_x)
                            if deltax1 < deltax2:
                                cat_x_center = c1cat
                                cat_x_side = c2cat
                            elif deltax2 < deltax1:
                                cat_x_center = c2cat
                                cat_x_side = c1cat
                            else:
                                flag_x_equal = True
                            
                            if c1_z > c2_z:
                                cat_z_front = c1cat
                                cat_z_back = c2cat
                            elif c2_z > c1_z:
                                cat_z_front = c2cat
                                cat_z_back = c1cat
                            else:
                                flag_z_equal = True
                                
                        elif orient == 'ToRight':
                            deltax1 = abs(c1_z - p_z)
                            deltax2 = abs(c2_z - p_z)
                            if deltax1 < deltax2:
                                cat_x_center = c1cat
                                cat_x_side = c2cat
                            elif deltax2 < deltax1:
                                cat_x_center = c2cat
                                cat_x_side = c1cat
                            else:
                                flag_x_equal = True
                            
                            if c1_x > c2_x:
                                cat_z_front = c1cat
                                cat_z_back = c2cat
                            elif c2_x > c1_x:
                                cat_z_front = c2cat
                                cat_z_back = c1cat
                            else:
                                flag_z_equal = True
                            
                        elif orient == 'ToLeft':
                            deltax1 = abs(c1_z - p_z)
                            deltax2 = abs(c2_z - p_z)
                            if deltax1 < deltax2:
                                cat_x_center = c1cat
                                cat_x_side = c2cat
                            elif deltax2 < deltax1:
                                cat_x_center = c2cat
                                cat_x_side = c1cat
                            else:
                                flag_x_equal = True
                                
                            if c1_x < c2_x:
                                cat_z_front = c1cat
                                cat_z_back = c2cat
                            elif c2_x < c1_x:
                                cat_z_front = c2cat
                                cat_z_back = c1cat
                            else:
                                flag_z_equal = True
                            
                        elif orient == 'Round':
                            # no medium infor for round table                            
                            dist1 = math.sqrt((c1_x-p_x)*(c1_x-p_x) + (c1_z-p_z)*(c1_z-p_z))
                            dist2 = math.sqrt((c2_x-p_x)*(c2_x-p_x) + (c2_z-p_z)*(c2_z-p_z))
                            if dist1 > dist2:
                                cat_z_front = c1cat
                                cat_z_back = c2cat
                            elif dist2 > dist1:
                                cat_z_front = c2cat
                                cat_z_back = c1cat
                            else:
                                flag_z_equal = True
                            
                        elif orient == 'Rectangle':
                            deltax1 = abs(c1_x-p_x)
                            deltaz1 = abs(c1_z-p_z)
                            deltax2 = abs(c2_x-p_x)
                            deltaz2 = abs(c2_z-p_z)
                            deltadeltax = abs(deltax1 - deltax2)
                            deltadeltaz = abs(deltaz1 - deltaz2)
                            if (deltadeltax > deltadeltaz): # use x to judge
                                if deltax1 < deltax2:
                                    cat_x_center = c1cat # the closer the more medium
                                    cat_x_side = c2cat
                                    cat_z_front = c2cat # the further the more front
                                    cat_z_back = c1cat
                                elif deltax2 < deltax1:
                                    cat_x_center = c2cat
                                    cat_x_side = c1cat
                                    cat_z_front = c1cat
                                    cat_z_back = c2cat
                                else:
                                    flag_x_equal = True
                                    flag_z_equal = True
                            else: # deltadeltax < deltadeltaz # use z to judge
                                if deltaz1 < deltaz2:
                                    cat_x_center = c1cat
                                    cat_x_side = c2cat
                                    cat_z_front = c2cat
                                    cat_z_back = c1cat
                                elif deltaz2 < deltaz1:
                                    cat_x_center = c2cat
                                    cat_x_side = c1cat
                                    cat_z_front = c1cat
                                    cat_z_back = c2cat
                                else:
                                    flag_x_equal = True
                                    flag_z_equal = True
                                
                            
                        elif orient == 'Object':
                            a = 1
                        elif orient == 'HangingObject':
                            a = 1
                        else:
                            a = 1

                    if flag_x_equal:
                        pair = (c1cat,c2cat) if c1cat < c2cat else (c2cat,c1cat)
                        if pair not in medium_equal_preferences:
                            medium_equal_preferences[pair] = 1
                        else:
                            medium_equal_preferences += 1
                    else:
                        pair = (cat_x_center, cat_x_side)
                        if pair not in medium_center_preferences:
                            medium_center_preferences[pair] = 1
                        else:
                            medium_center_preferences[pair] += 1
                    if flag_z_equal:
                        pair = (c1cat,c2cat) if c1cat < c2cat else (c2cat,c1cat)
                        if pair not in depth_equal_preferences:
                            depth_equal_preferences[pair] = 1
                        else:
                            depth_equal_preferences[pair] += 1
                    else:
                        pair = (cat_z_front, cat_z_back)
                        if pair not in depth_front_preferences:
                            depth_front_preferences[pair] = 1
                        else:
                            depth_front_preferences[pair] += 1
                        
                    
        
print len(height_higher_preferences)
print len(height_equal_preferences)
print (medium_center_preferences)
print len(medium_equal_preferences)
print (depth_front_preferences)
print len(depth_equal_preferences)

   
   
```
