```python
from os import walk
scenepath='./../scenes'

f = []
for (dirpath, dirnames, filenames) in walk(scenepath):      
    f.extend(filenames)
    break

sceneinfo = []

for f in filenames:    
    if f.endswith('.txt'):
        sceneinfo.append(f)
        sceneinfo.append('\n')
        with open(scenepath + '/' + f) as fp:            
            for line in fp:
                if line.startswith('newModel'):
                    modelid = line.split(' ')[-1]
                    modelIndex = line.split(' ')[1]
                if line.startswith('children'):
                    childnum = len(line.split(' ')) - 1
                    if childnum > 0: # record this model                        
                        sceneinfo.append(modelIndex + '\t' + modelid)               
        sceneinfo.append('\n')

                    
                    
                    
                    
```

```python
# save info                
with open('./ParentObjFrontDirection.txt', 'w') as writer:
    for item in sceneinfo:
        writer.write("%s" % item)
                    
                    
```
